```python
# 라이브러리 불러오기
import pandas as pd
```

```python
# 데이터 불러오기
train = pd.read_csv("../input/big-data-analytics-certification/t2-1-train.csv")
test = pd.read_csv("../input/big-data-analytics-certification/t2-1-test.csv")
```

```python
train['AnnualIncome'] = train['AnnualIncome'].fillna(train['AnnualIncome'].mean())
test['AnnualIncome'] = test['AnnualIncome'].fillna(test['AnnualIncome'].mean())
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

train_cols = train.select_dtypes(include = 'object').columns
test_cols = test.select_dtypes(include = 'object').columns

for cols in train_cols:
    train[cols] = encoder.fit_transform(train[cols])
    
for cols in test_cols:
    test[cols] = encoder.fit_transform(test[cols])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
train.head()
```

```python
target = train.pop('TravelInsurance')
```

```python
train['AnnualIncome'] = scaler.fit_transform(train[['AnnualIncome']])
test['AnnualIncome'] = scaler.transform(test[['AnnualIncome']])
test
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size = 0.15, random_state = 42)
X_tr.shape, X_val.shape, y_tr.shape, y_val.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import roc_auc_score
```

```python
model = LogisticRegression()
model.fit(X_tr, y_tr)
pred = model.predict(X_val)
a1 = roc_auc_score(y_val, pred)
a1
```

```python
model = RandomForestClassifier()
model.fit(X_tr, y_tr)
pred = model.predict(X_val)
a2 = roc_auc_score(y_val, pred)
a2
```

```python
model = DecisionTreeClassifier()
model.fit(X_tr, y_tr)
pred = model.predict(X_val)
a3 = roc_auc_score(y_val, pred)
a3
```
