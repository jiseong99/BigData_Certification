## 문제1
- 데이터셋(basic1.csv)의 'f5' 컬럼을 기준으로 상위 10개의 데이터를 구하고,
- 'f5'컬럼 10개 중 최소값으로 데이터를 대체한 후, 
- 'age'컬럼에서 80 이상인 데이터의'f5 컬럼 평균값 구하기

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd

df = pd.read_csv('../input/bigdatacertificationkr/basic1.csv')
df.head()
```

```python
df = df.sort_values('f5', ascending = False)
df
```

```python
min_f5 = df['f5'].head(10).min()
min_f5
```

```python
df.iloc[:10, -1] = min_f5
df.head(11)
```

```python
# 80세 이상의 f5컬럼 평균
print(df[df['age']>=80]['f5'].mean())
```

## 문제2
- 데이터셋(basic1.csv)의 앞에서 순서대로 70% 데이터만 활용해서,
- 'f1'컬럼 결측치를 중앙값으로 채우기 전후의 표준편차를 구하고
- 두 표준편차 차이 계산하기

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
import numpy as np

df = pd.read_csv('../input/bigdatacertificationkr/basic1.csv')
```

```python
df.head()
```

```python
len(df)
```

```python
df70 = df.iloc[0:70]
df30 = df.iloc[70:30]
df70
```

```python
ans1 = df70['f1'].std()
ans1
```

```python
med = df70.f1.median()
med
```

```python
df70['f1'] = df70['f1'].fillna(med)
```

```python
ans2 = df70['f1'].std()
ans2
```

```python
print(ans1 - ans2)
```

## 문제3
- 데이터셋(basic1.csv)의 'age'컬럼의 이상치를 더하시오!
- 단, 평균으로부터 '표준편차*1.5'를 벗어나는 영역을 이상치라고 판단함

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
import numpy as np

df = pd.read_csv('../input/bigdatacertificationkr/basic1.csv')
```

```python
df.head()
```

```python
mn = df['age'].mean()
stand = df['age'].std()
```

```python
a = mn + 1.5 * stand
b = mn - 1.5 * stand
a, b
```

```python
df = df.loc[(df['age'] < b) | (df['age'] > a)]
df
```
