## 주어진 데이터에서 2022년 5월 주말과 평일의 sales컬럼 평균값 차이를 구하시오 (소수점 둘째자리까지 출력, 반올림)
- 데이터셋 : basic2.csv
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작

```python
# 라이브러리 불러오기
import pandas as pd
```

```python
# 데이터 불러오기 (datatime컬럼 지정)
df = pd.read_csv("../input/bigdatacertificationkr/basic2.csv", parse_dates=['Date'])
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['dayofweek'] = df['Date'].dt.dayofweek

df.head()
```

```python
weekend = df[(df['dayofweek'] == 5) | (df['dayofweek'] == 6)]
weekday = df[(df['dayofweek'] != 5) & (df['dayofweek'] != 6)]
```

```python
weekend.head(10)
```

```python
weekday.head(10)
```

```python
day = weekday.loc[(weekday['year'] == 2022) & (weekday['month'] == 5)]
end = weekend.loc[(weekend['year'] == 2022) & (weekend['month'] == 5)]
```

```python
end_mean = end['Sales'].mean()
day_mean = day['Sales'].mean()
```

```python
print(round(end_mean - day_mean, 1))
```
