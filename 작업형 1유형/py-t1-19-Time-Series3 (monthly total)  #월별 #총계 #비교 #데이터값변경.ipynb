### 주어진 데이터에서 2022년 월별 Sales 합계 중 가장 큰 금액과 
### 2023년 월별 Sales 합계 중 가장 큰 금액의 차이를 절대값으로 구하시오. 
### 단 Events컬럼이 '1'인경우 80%의 Salse값만 반영함 
### (최종값은 소수점 반올림 후 정수 출력)

```python
# 라이브러리 불러오기
import pandas as pd
```

```python
# 데이터 불러오기 (datatime컬럼 지정)
df = pd.read_csv("../input/bigdatacertificationkr/basic2.csv", 
                 parse_dates=['Date'])
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df
```

```python
df.loc[df['Events'] == 1, 'Sales'] *= 0.8
```

```python
df_2022 = df.loc[df['year'] == 2022]
df_2023 = df.loc[df['year'] == 2023]
```

```python
df_2022.head(2)
```

```python
m22 = df_2022.groupby(df['month']).sum()
m22
```

```python
m23 = df_2023.groupby(df['month']).sum()
m23
```

```python
print(abs(int(round(m22['Sales'].max() - m23['Sales'].max(),0))))
```
